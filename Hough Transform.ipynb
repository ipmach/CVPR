# Hough Line detection

p = x cos(theta) + y sin(theta)

## Load image

```python
img = cv2.imread("img/lines_with_box.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

## Do edge detection

```python
edge_img = CannyEdge.apply(img)

plt.imshow(edge_img, cmap='gray')
```

## Apply hough

```python
top_, accumulator, accumulator2 = HoughTransform.apply(edge_img, plot=False, num_lines=5)
```

### Accumulator

```python
plt.imshow(accumulator, cmap='hot')
```

### Accumulator maxima supresion

```python
plt.imshow(accumulator2, cmap='hot')
```

### Drawing Lines

```python
HoughTransform.draw_line(img, top_)
```

# Hough circle detection

(x - a)^2 + (y-b)^2 = r^2

## Load image

```python
img = cv2.imread("img/watch.pgm", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

## Do edge detection

```python
edge_img = CannyEdge.apply(img)

plt.imshow(edge_img, cmap='gray')
```

## Apply Hough

```python
top_, accumulator, accumulator2 = HoughTransform.apply(edge_img, plot=True, 
                                                       num_lines=2, r=60, method='circle')
```

### Accumulator

```python
plt.imshow(accumulator, cmap='hot')
```

### Accumulator maxima supresion

```python
plt.imshow(accumulator2, cmap='hot')
```
