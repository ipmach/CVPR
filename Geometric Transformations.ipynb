# Load dependencies

```python
%cd CVPR
from ComputerVision.Geometric_Transformations.geometric_matrix import GeometricTransformations
from ComputerVision.Geometric_Transformations.mappings import inverse_mapping, forward_mapping
from ComputerVision.Geometric_Transformations.interpolations import Interpolations
import matplotlib.pyplot as plt
import numpy as np
import cv2
%cd ..
```

# Load image

```python
img = cv2.imread("img/watch.pgm", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

# Geometric transformations

Different examples with inverse and forward mapping

## Scaling

Scaling by inverse and by forward mapping

### Scaling by forward

```python
A = GeometricTransformations.scaling_k(2,1)
img2 = forward_mapping(img, A, scale=(2, 1))
plt.imshow(img2, cmap='gray')
```

### Scaling by inverse

```python
A = GeometricTransformations.scaling_k(2,1)
img3 = inverse_mapping(img, A, scale=(2, 1))
plt.imshow(img3, cmap='gray')
```

### Difference

```python
plt.imshow(abs(img3 - img2))
plt.colorbar()
```
