# Load dependencies

```python
%cd CVPR
from ComputerVision.Values_Transformations.intensities import Intensities
import matplotlib.pyplot as plt
import numpy as np
import cv2
%cd ..
```

# Load image

```python
img = cv2.imread("img/dog.pgm", cv2.IMREAD_GRAYSCALE)
intensity = np.arange(100) / 100

plt.figure(figsize=(11, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("Intensity vs Intensity")
plt.plot(intensity, intensity)
plt.xlabel("I")
plt.ylabel("I")
```

# Intensity Transformations

The different intensity transformations work with Intensity between [0,1]

## Filter Transformation

### Apply filter

```python
img_p = img / 255
img2 = Intensities.filter_transform_v(img_p, r0=0.2, r2=0.7)
TI = Intensities.filter_transform_v(intensity, r0=0.2, r2=0.7)

plt.figure(figsize=(11, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2 * 255, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("Intensity vs Intensity")
plt.plot(intensity, TI)
plt.xlabel("I")
plt.ylabel("T(I)")
```

### Apply filter binary

```python
img_p = img / 255
img2 = Intensities.filter_transform_v(img_p, r0=0.2, r2=0.7, binary=True)
TI = Intensities.filter_transform_v(intensity, r0=0.2, r2=0.7, binary=True)

plt.figure(figsize=(11, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2 * 255, cmap='gray')
plt.subplot(1, 2, 2)
plt.title("Intensity vs Intensity")
plt.plot(intensity, TI)
plt.xlabel("I")
plt.ylabel("T(I)")
```
