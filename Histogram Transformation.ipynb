# Load image

```python
img = cv2.imread("img/dog.pgm", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
```

### Image Histogram

```python
values, count = Histogram.histogram(img, figsize=(13, 4))
```

### Image Histogram Normalize

```python
values, count = Histogram.histogram(img, figsize=(13, 4), normalize=True)
```

# Histogram Transformation

## Histogram Equalization

Different methods comparison

### Book's method

```python
img2 = Histogram.equalization(img, method='book')
plt.imshow(img2, cmap='gray')
values, count = Histogram.histogram(img2, figsize=(13, 4))
```

### Wikipedia's method

```python
img2 = Histogram.equalization(img, method='wiki')
plt.imshow(img2, cmap='gray')
values, count = Histogram.histogram(img2, figsize=(13, 4))
```

### Profesor's method

```python
img2 = Histogram.equalization(img, method='prof')
plt.imshow(img2, cmap='gray')
values, count = Histogram.histogram(img2, figsize=(13, 4))
```

## Histogram Matching

## Load matching image

```python
img2 = cv2.imread("img/forest.pgm", cv2.IMREAD_GRAYSCALE)
plt.imshow(img2, cmap='gray')
values, count = Histogram.histogram(img2, figsize=(13, 4))
```

## Applying matching

```python
img3 = Histogram.matching(img, img2)
plt.imshow(img3, cmap='gray')
values, count = Histogram.histogram(img3, figsize=(13, 4))
```
